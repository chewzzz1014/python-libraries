```python
path = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv"
```

```python
import pandas as pd

df = pd.read_csv(path)
df
```

```python
def calc_substrate_in_rows(df, col):
    num_rows = df[col].shape[0]
    
    results = [df[col][0]]
    for i in range(1, num_rows):
        results.append(df[col][i] - df[col][i-1])
    
    df["substrate"] = pd.Series(results)
    
    return df
```

```python
df = calc_substrate_in_rows(df, "mpg")
df
```
