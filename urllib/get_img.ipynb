Download many images from the links

```python
import os
from urllib.request import urlretrieve
import matplotlib.pyplot as plt
```

```python
os.listdir(".")
```

```python
path_template = "https://raw.githubusercontent.com/chewzzz1014/Web-Dev-For-Beginners/main/3-terrarium/solution/images/"
dest_template = "./images/plant"

for i in range(1, 15):
    path = path_template + "plant" + str(i) + ".png"
    dest = dest_template + str(i) + ".png"
    urlretrieve(path, dest)
    
```

```python
# downloaded successfully
from PIL import Image
img = Image.open("./images/plant1.png") 

plt.axis("off")
plt.imshow(img);
```
