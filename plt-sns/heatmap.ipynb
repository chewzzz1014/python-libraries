```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset("flights").pivot("month", "year", "passengers")
df
```

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(df);
```

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(df, fmt="d", annot=True, cmap="Blues");
```
