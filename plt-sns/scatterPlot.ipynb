## Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset("iris")
df
```

```python
# only has 3 unique species
print(df.species.unique())
```

```python
# using line plot
plt.plot(df.sepal_length, df.sepal_width);
# what a mess!
```

```python
sns.scatterplot(x=df["sepal_length"], y=df["sepal_width"]);
```

## Adding Hues
- Form clusters with different stylings

```python
plt.figure(figsize=(12, 6))

plt.title("Sepal Dimensions")

sns.set_style("whitegrid")
sns.scatterplot(x=df.sepal_length, y=df.sepal_width, hue=df.species, s=100);
```

## Plotting Using Pandas Data Frames (dataframe as argument)

```python
plt.title("Sepal Dimensions")

sns.scatterplot(x="sepal_length", y="sepal_width",
               hue="species", s=100, data=df);
```
