## Line Chart, plt.plot(data_x, data_y)
- plt.plot(data_y): When only one argument is given
- plt.plot(data_x, data_y): 2 arguments, each for axis-x and axis-y respectively

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(yield_apples)
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(years, yield_apples)

# axis labels
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title("Apple Yields in 2022");
```

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]

# multiple lines in one plot
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title("Apples and Oranges Yield from 2022 to 2012")

# legend
plt.legend(["Apples", "Oranges"]);
```

## Styling and Markers
**plt.plot supports many arguments for styling lines and markers:**
- `color` or `c`: Color of line
- `linestyle` or `ls`: Solid or dashed lins
- `linewidth` or `lw`: Width of line
- `markersize` or `ms`: Size of markers
- `markeredgecolor` or `mec`: Edge color for markers
- `markeredgewidth` or `mew`: Edge width for markers
- `markerfacecolor` or `mfc`: Fill color for markers
- `alpha`: Opacity of plot
- `[marker][line][color]`: Shorthand for specifying the marker shape, line style, and line color. 

```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=0.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

## Changing Figure Size, plt.figure(figsize=(width, height))

```python
plt.figure(figsize=(12, 6))

plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```
