## Use Matplotlib to Display Images

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
from urllib.request import urlretrieve

urlretrieve("https://i.imgur.com/SkPbq.jpg", "chart.jpg");
```

```python
# read image into memory using PIL module
from PIL import Image

# img is just numpy array contain pixel intensities for RGB
img = Image.open("chart.jpg")

img_arr = np.array(img)
img_arr.shape
```

```python
# show the image
plt.imshow(img);
```

```python
# turn off gird lines
plt.grid(False)

# turn off axes
plt.axis("off")

plt.title("A meme")
plt.imshow(img);
```

```python
# displaying a pary of image only
plt.grid(False)
plt.axis('off')
plt.imshow(img_arr[125:325,105:305]);
```
